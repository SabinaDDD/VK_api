```python
users_df = pd.read_excel('/content/drive/MyDrive/vk_users.xlsx', index_col='ID')
users_ids = users_df.index
print(users_df.head())
print(users_ids)
```

```python
url_user = users_df.loc[33335, 'URL'] # пробуем вытащить адрес конкретного юзера 
name = users_df.loc[33335,'Имя'] + ' ' +users_df.loc[33335,'Фамилия'] # пробуем вытащить имя и фамилию того же юзера 
print(url_user)
print(name)
```

```python
data = []
for user_id in users_ids: # пройдемся по всем id пользователей 
  params = {'access_token': token, 'v': 5.95, 'owner_id': user_id, 'count': 1, 'filter': 'owner'} # подставляем айди каждого польвователя в словарь с параментрами 
  response = requests.get(url, params = params) # делаем запрос 
  json = response.json() # приводим к структуре json
  
  if 'response' in json.keys() and json['response']['count'] > 0 and 'copy_history' not in json['response']['items'][0].keys(): #проверяем на наличие ответа как такового 
    dt_unix = json['response']['items'][0]['date'] # берем дату поста 
    dt = datetime.utcfromtimestamp(dt_unix).strftime('%Y-%m-%d') # преобразовываем к формату datetime
    if dt >= '2022-04-01': # ставим условие чтобы дата была не ранее 1 апреля 2022 года 
      post_text = json['response']['items'][0]['text'].rstrip().strip() # берем текст самого поста 
      url_user = users_df.loc[user_id, 'URL'] # из имеющейс таблицы берем адрес юзера  
      name = users_df.loc[user_id,'Имя'] + ' ' +users_df.loc[user_id,'Фамилия'] # также берем имя и фамилию
      new = [user_id, url_user, name,  dt, post_text] # дклаем из всего этого список 
      data.append(new) # добавляем в общий список 
  if len(data) > 100:
    break
```

```python
pprint(data)
```

```python
data_df = pd.DataFrame(data = data, columns=['user_id', 'url', 'name', 'datetime', 'post_text'])
print(data_df.head())
```
