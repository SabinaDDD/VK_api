```python
users_df = pd.read_excel('/content/drive/MyDrive/vk_users.xlsx', index_col='ID')
users_ids = users_df.index
print(users_df.head())
print(users_ids)
```

```python
url_user = users_df.loc[33335, 'URL'] # –ø—Ä–æ–±—É–µ–º –≤—ã—Ç–∞—â–∏—Ç—å –∞–¥—Ä–µ—Å –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–≥–æ —é–∑–µ—Ä–∞ 
name = users_df.loc[33335,'–ò–º—è'] + ' ' +users_df.loc[33335,'–§–∞–º–∏–ª–∏—è'] # –ø—Ä–æ–±—É–µ–º –≤—ã—Ç–∞—â–∏—Ç—å –∏–º—è –∏ —Ñ–∞–º–∏–ª–∏—é —Ç–æ–≥–æ –∂–µ —é–∑–µ—Ä–∞ 
print(url_user)
print(name)
```

```python
data = []
for user_id in users_ids: # –ø—Ä–æ–π–¥–µ–º—Å—è –ø–æ –≤—Å–µ–º id –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π 
  params = {'access_token': token, 'v': 5.95, 'owner_id': user_id, 'count': 1, 'filter': 'owner'} # –ø–æ–¥—Å—Ç–∞–≤–ª—è–µ–º –∞–π–¥–∏ –∫–∞–∂–¥–æ–≥–æ –ø–æ–ª—å–≤–æ–≤–∞—Ç–µ–ª—è –≤ —Å–ª–æ–≤–∞—Ä—å —Å –ø–∞—Ä–∞–º–µ–Ω—Ç—Ä–∞–º–∏ 
  response = requests.get(url, params = params) # –¥–µ–ª–∞–µ–º –∑–∞–ø—Ä–æ—Å 
  json = response.json() # –ø—Ä–∏–≤–æ–¥–∏–º –∫ —Å—Ç—Ä—É–∫—Ç—É—Ä–µ json
  
  if 'response' in json.keys() and json['response']['count'] > 0 and 'copy_history' not in json['response']['items'][0].keys(): #–ø—Ä–æ–≤–µ—Ä—è–µ–º –Ω–∞ –Ω–∞–ª–∏—á–∏–µ –æ—Ç–≤–µ—Ç–∞ –∫–∞–∫ —Ç–∞–∫–æ–≤–æ–≥–æ 
    dt_unix = json['response']['items'][0]['date'] # –±–µ—Ä–µ–º –¥–∞—Ç—É –ø–æ—Å—Ç–∞ 
    dt = datetime.utcfromtimestamp(dt_unix).strftime('%Y-%m-%d') # –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤—ã–≤–∞–µ–º –∫ —Ñ–æ—Ä–º–∞—Ç—É datetime
    if dt >= '2022-04-01': # —Å—Ç–∞–≤–∏–º —É—Å–ª–æ–≤–∏–µ —á—Ç–æ–±—ã –¥–∞—Ç–∞ –±—ã–ª–∞ –Ω–µ —Ä–∞–Ω–µ–µ 1 –∞–ø—Ä–µ–ª—è 2022 –≥–æ–¥–∞ 
      post_text = json['response']['items'][0]['text'].rstrip().strip() # –±–µ—Ä–µ–º —Ç–µ–∫—Å—Ç —Å–∞–º–æ–≥–æ –ø–æ—Å—Ç–∞ 
      url_user = users_df.loc[user_id, 'URL'] # –∏–∑ –∏–º–µ—é—â–µ–π—Å —Ç–∞–±–ª–∏—Ü—ã –±–µ—Ä–µ–º –∞–¥—Ä–µ—Å —é–∑–µ—Ä–∞  
      name = users_df.loc[user_id,'–ò–º—è'] + ' ' +users_df.loc[user_id,'–§–∞–º–∏–ª–∏—è'] # —Ç–∞–∫–∂–µ –±–µ—Ä–µ–º –∏–º—è –∏ —Ñ–∞–º–∏–ª–∏—é
      new = [user_id, url_user, name,  dt, post_text] # –¥–∫–ª–∞–µ–º –∏–∑ –≤—Å–µ–≥–æ —ç—Ç–æ–≥–æ —Å–ø–∏—Å–æ–∫ 
      data.append(new) # –¥–æ–±–∞–≤–ª—è–µ–º –≤ –æ–±—â–∏–π —Å–ø–∏—Å–æ–∫ 
  if len(data) > 100:
    break
```
